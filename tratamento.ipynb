## Geração dos INSERTs para o Banco de Dados

```python
# Função auxiliar para gerar INSERT de dimensões simples
def gerar_insert_dimensao(tabela, coluna, valores_unicos):
    """Gera INSERT para tabelas de dimensão simples"""
    inserts = []
    inserts.append(f"-- Inserir dados na tabela {tabela}")
    
    for valor in valores_unicos:
        if pd.notna(valor):  # Ignorar valores nulos
            valor_escaped = str(valor).replace("'", "''")
            inserts.append(f"INSERT INTO epidemiologia.{tabela} ({coluna}) VALUES ('{valor_escaped}') ON CONFLICT DO NOTHING;")
    
    return "\n".join(inserts)

# Função para gerar INSERT de dimensões booleanas
def gerar_insert_booleano(tabela):
    """Gera INSERT para tabelas booleanas"""
    inserts = []
    inserts.append(f"-- Inserir dados na tabela {tabela}")
    inserts.append(f"INSERT INTO epidemiologia.{tabela} (status) VALUES (TRUE) ON CONFLICT DO NOTHING;")
    inserts.append(f"INSERT INTO epidemiologia.{tabela} (status) VALUES (FALSE) ON CONFLICT DO NOTHING;")
    return "\n".join(inserts)

print("Funções auxiliares criadas!")
```

```python
# Gerar INSERTs para todas as dimensões simples
sql_script = []
sql_script.append("-- =========================================")
sql_script.append("-- SCRIPT DE INSERÇÃO DE DADOS")
sql_script.append("-- =========================================\n")

# 1. Sexo
sql_script.append(gerar_insert_dimensao("sexo", "nome", df["sexo"].unique()))
sql_script.append("")

# 2. Raça
sql_script.append(gerar_insert_dimensao("raca", "nome", df["racacor"].unique()))
sql_script.append("")

# 3. Evolução
sql_script.append(gerar_insert_dimensao("evolucao", "nome", df["evolucaocaso"].unique()))
sql_script.append("")

# 4. Classificação
sql_script.append(gerar_insert_dimensao("classificacao", "nome", df["classificacaofinal"].unique()))
sql_script.append("")

# 5. Estratégia COVID
sql_script.append(gerar_insert_dimensao("estrategia_covid", "codigo", df["codigoestrategiacovid"].unique()))
sql_script.append("")

# 6. Resultado Teste 1
sql_script.append(gerar_insert_dimensao("resultado_teste1", "codigo", df["codigoresultadoteste1"].unique()))
sql_script.append("")

# 7. Profissional Saúde (booleano)
sql_script.append(gerar_insert_booleano("profissional_saude"))
sql_script.append("")

# 8. Profissional Segurança (booleano)
sql_script.append(gerar_insert_booleano("profissional_seguranca"))
sql_script.append("")

print("INSERTs de dimensões simples gerados!")
print(f"Total de linhas SQL: {len(sql_script)}")
```

```python
# Mapear siglas para códigos IBGE de estados brasileiros
codigos_ibge_estados = {
    'ac': 12, 'al': 27, 'ap': 16, 'am': 13, 'ba': 29, 'ce': 23,
    'df': 53, 'es': 32, 'go': 52, 'ma': 21, 'mt': 51, 'ms': 50,
    'mg': 31, 'pa': 15, 'pb': 25, 'pr': 41, 'pe': 26, 'pi': 22,
    'rj': 33, 'rn': 24, 'rs': 43, 'ro': 11, 'rr': 14, 'sc': 42,
    'sp': 35, 'se': 28, 'to': 17
}

# Gerar INSERTs para Estados
sql_script.append("-- Inserir dados na tabela estado")

# Criar conjunto de siglas únicas
siglas_unicas = set()
if "estadoibge" in df.columns:
    siglas_unicas.update(df["estadoibge"].dropna().unique())
if "estadonotificacaoibge" in df.columns:
    siglas_unicas.update(df["estadonotificacaoibge"].dropna().unique())

# Gerar INSERTs
for sigla in sorted(siglas_unicas):
    sigla_lower = str(sigla).lower().strip()
    if sigla_lower in codigos_ibge_estados:
        codigo_ibge = codigos_ibge_estados[sigla_lower]
        sigla_upper = sigla_lower.upper()
        sql_script.append(f"INSERT INTO epidemiologia.estado (sigla, codigo_ibge) VALUES ('{sigla_upper}', {codigo_ibge}) ON CONFLICT DO NOTHING;")

sql_script.append("")
print(f"Estados únicos: {len(siglas_unicas)}")
print("INSERTs de estados gerados!")
```

```python
# Gerar INSERTs para Municípios
sql_script.append("-- Inserir dados na tabela municipio")

# Coletar municípios de origem
municipios = []
if "municipio" in df.columns and "municipioibge" in df.columns and "estadoibge" in df.columns:
    mun_df = df[["municipio", "municipioibge", "estadoibge"]].copy()
    mun_df = mun_df.dropna().drop_duplicates()
    
    for _, row in mun_df.iterrows():
        nome = str(row["municipio"]).replace("'", "''").title()
        codigo_ibge = str(row["municipioibge"]).strip()
        sigla_estado = str(row["estadoibge"]).upper().strip()
        
        # Validar código IBGE numérico
        try:
            codigo_ibge_int = int(float(codigo_ibge))
            municipios.append((nome, codigo_ibge_int, sigla_estado))
        except:
            pass

# Coletar municípios de notificação
if "municipionotificacao" in df.columns and "municipionotificacaoibge" in df.columns and "estadonotificacaoibge" in df.columns:
    mun_notif_df = df[["municipionotificacao", "municipionotificacaoibge", "estadonotificacaoibge"]].copy()
    mun_notif_df = mun_notif_df.dropna().drop_duplicates()
    
    for _, row in mun_notif_df.iterrows():
        nome = str(row["municipionotificacao"]).replace("'", "''").title()
        codigo_ibge = str(row["municipionotificacaoibge"]).strip()
        sigla_estado = str(row["estadonotificacaoibge"]).upper().strip()
        
        try:
            codigo_ibge_int = int(float(codigo_ibge))
            municipios.append((nome, codigo_ibge_int, sigla_estado))
        except:
            pass

# Remover duplicatas
municipios_unicos = list(set(municipios))

# Gerar INSERTs
for nome, codigo_ibge, sigla_estado in sorted(municipios_unicos, key=lambda x: x[1]):
    sql_script.append(f"INSERT INTO epidemiologia.municipio (nome, codigo_ibge, estado_id) ")
    sql_script.append(f"SELECT '{nome}', {codigo_ibge}, estado_id FROM epidemiologia.estado WHERE sigla = '{sigla_estado}' ")
    sql_script.append(f"ON CONFLICT DO NOTHING;")

sql_script.append("")
print(f"Municípios únicos: {len(municipios_unicos)}")
print("INSERTs de municípios gerados!")
```

```python
# Salvar primeiro o script de dimensões
with open("inserts_dimensoes.sql", "w", encoding="utf-8") as f:
    f.write("\n".join(sql_script))

print(f"Arquivo 'inserts_dimensoes.sql' criado com sucesso!")
print(f"Total de linhas: {len(sql_script)}")
```
