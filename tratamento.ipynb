```python
import pandas as pd
```

```python
df = pd.read_csv("dataseet-isaac.csv", encoding="utf-8")
```

```python
limite = len(df) * 0.40  
df = df.dropna(thresh=limite, axis=1)
```

```python
#esse trecho aqui normaliza o banco que tem muita letra maiuscula e minuscula ai normaliza tudo pra minuscula

for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.lower().str.strip()
```

```python
df.columns
```

```python
df.columns = (
    df.columns
    .str.lower()          
    .str.replace(" ", "") 
    .str.replace("-", "") 
    .str.replace("/", "") 
)
# nornalizei as tabelas que estavam todas cagadas tirando virgula,barra e a porra toda
```

```python
df.columns.tolist()
```

```python
map_raca = {
    "branca": "branca",
    "branco": "branca",
    "preta": "preta",
    "negra": "preta",
    "parda": "parda",
    "amarela": "amarela",
    "indigena": "indigena"
}

df["racacor"] = df["racacor"].str.lower().map(map_raca)
df["racacor"] = df["racacor"].fillna("ignorado")
```

```python
map_sexo = {
    "m": "masculino",
    "masculino": "masculino",
    "f": "feminino",
    "feminino": "feminino"
}

df["sexo"] = df["sexo"].str.lower().map(map_sexo)
df["sexo"] = df["sexo"].fillna("ignorado")
```

```python
map_prof_saude = {
    "sim": 1,
    "não": 0,
    "nao": 0,
    "n": 0,
    "s": 1
}

df["profissionalsaude"] = df["profissionalsaude"].str.lower().map(map_prof_saude)
df["profissionalsaude"] = df["profissionalsaude"].fillna(0)
```

```python
map_prof_seg = {
    "sim": 1,
    "não": 0,
    "nao": 0,
    "n": 0,
    "s": 1
}

df["profissionalseguranca"] = df["profissionalseguranca"].str.lower().map(map_prof_seg)
df["profissionalseguranca"] = df["profissionalseguranca"].fillna(0)
```

```python
map_evolucao = {
    "cura": "cura",
    "óbito": "obito",
    "obito": "obito",
    "internado": "internado",
    "hospitalizado": "internado",
    "ignorado": "ignorado"
}

df["evolucaocaso"] = df["evolucaocaso"].str.lower().map(map_evolucao)
df["evolucaocaso"] = df["evolucaocaso"].fillna("ignorado")
```

```python
map_classif = {
    "confirmado": "confirmado",
    "descartado": "descartado",
    "suspeito": "suspeito"
}

df["classificacaofinal"] = df["classificacaofinal"].str.lower().map(map_classif)
df["classificacaofinal"] = df["classificacaofinal"].fillna("ignorado")
```

```python
##tratar coluna de datas
date_cols = [
    "datanotificacao", "datainiciosintomas", "dataencerramento",
    "dataprimeiradose", "datasegundadose",
    "datacoletateste1", "datacoletateste2",
    "datacoletateste3", "datacoletateste4"
]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
```

```python
df["idade"] = pd.to_numeric(df["idade"], errors="coerce")
```

```python
df["raw_line"] = df.astype(str).apply(lambda x: ";".join(x), axis=1)
```

```python
df.to_csv("dataset_limpo.csv", index=False, encoding="utf-8")
```
