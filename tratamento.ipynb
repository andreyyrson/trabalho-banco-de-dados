```python
import pandas as pd
```

```python
df = pd.read_csv("dataseet-isaac.csv", encoding="utf-8")
```

```python
limite = len(df) * 0.40  
df = df.dropna(thresh=limite, axis=1)
```

```python
#esse trecho aqui normaliza o banco que tem muita letra maiuscula e minuscula ai normaliza tudo pra minuscula

for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.lower().str.strip()
```

```python
df.columns
```

```python
df.columns = (
    df.columns
    .str.lower()          
    .str.replace(" ", "") 
    .str.replace("-", "") 
    .str.replace("/", "") 
)
# nornalizei as tabelas que estavam todas cagadas tirando virgula,barra e a porra toda
```

```python
df.columns.tolist()
```

```python
map_raca = {
    "branca": "branca",
    "branco": "branca",
    "preta": "preta",
    "negra": "preta",
    "parda": "parda",
    "amarela": "amarela",
    "indigena": "indigena"
}

df["racacor"] = df["racacor"].str.lower().map(map_raca)
df["racacor"] = df["racacor"].fillna("ignorado")
```

```python
map_sexo = {
    "m": "masculino",
    "masculino": "masculino",
    "f": "feminino",
    "feminino": "feminino"
}

df["sexo"] = df["sexo"].str.lower().map(map_sexo)
df["sexo"] = df["sexo"].fillna("ignorado")
```

```python
map_prof_saude = {
    "sim": "sim",
    "não": "não",
    "nao": "não",
    "n": "não",
    "s": "sim"
}

df["profissionalsaude"] = df["profissionalsaude"].str.lower().map(map_prof_saude)
df["profissionalsaude"] = df["profissionalsaude"].fillna(0)
```

```python
map_prof_seg = {
    "sim": "sim",
    "não": "não",
    "nao": "não",
    "n": "não",
    "s": "sim"
}

df["profissionalseguranca"] = df["profissionalseguranca"].str.lower().map(map_prof_seg)
df["profissionalseguranca"] = df["profissionalseguranca"].fillna(0)
```

```python
map_evolucao = {
    "cura": "cura",
    "óbito": "obito",
    "obito": "obito",
    "internado": "internado",
    "hospitalizado": "internado",
    "ignorado": "ignorado"
}

df["evolucaocaso"] = df["evolucaocaso"].str.lower().map(map_evolucao)
df["evolucaocaso"] = df["evolucaocaso"].fillna("ignorado")
```

```python
map_classif = {
    "confirmado": "confirmado",
    "descartado": "descartado",
    "suspeito": "suspeito"
}

df["classificacaofinal"] = df["classificacaofinal"].str.lower().map(map_classif)
df["classificacaofinal"] = df["classificacaofinal"].fillna("ignorado")
```

```python
# Mapear códigos de estratégia COVID
# 1 = Sintomático, 2 = Contato, 3 = Testagem em massa
map_estrategia_covid = {
    1.0: "sintomatico",
    2.0: "contato",
    3.0: "testagem_massa"
}

if "codigoestrategiacovid" in df.columns:
    df["codigoestrategiacovid"] = df["codigoestrategiacovid"].map(map_estrategia_covid)
    df["codigoestrategiacovid"] = df["codigoestrategiacovid"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigoestrategiacovid"].value_counts())
else:
    print("Coluna 'codigoestrategiacovid' não encontrada no dataframe")
```

```python
# Mapear códigos de local de realização de testagem
# 1=Unidade de Saúde, 2=Domicílio, 3=Drive-thru, 4=Instituição de longa permanência
# 5=Instituição de ensino, 6=Presídio, 7=Outros
map_local_testagem = {
    1.0: "unidade_saude",
    2.0: "domicilio",
    3.0: "drive_thru",
    4.0: "instituicao_longa_permanencia",
    5.0: "instituicao_ensino",
    6.0: "presidio",
    7.0: "outros"
}

if "codigolocalrealizacaotestagem" in df.columns:
    df["codigolocalrealizacaotestagem"] = df["codigolocalrealizacaotestagem"].map(map_local_testagem)
    df["codigolocalrealizacaotestagem"] = df["codigolocalrealizacaotestagem"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigolocalrealizacaotestagem"].value_counts())
else:
    print("Coluna 'codigolocalrealizacaotestagem' não encontrada no dataframe")
```

```python
# Mapear códigos de vacina
# 1 = Sim, 2 = Não, 3 = Não se aplica
map_recebeu_vacina = {
    1.0: "sim",
    2.0: "nao",
    3.0: "nao_se_aplica"
}

if "codigorecebeuvacina" in df.columns:
    df["codigorecebeuvacina"] = df["codigorecebeuvacina"].map(map_recebeu_vacina)
    df["codigorecebeuvacina"] = df["codigorecebeuvacina"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigorecebeuvacina"].value_counts())
else:
    print("Coluna 'codigorecebeuvacina' não encontrada no dataframe")
```

```python
# Mapear códigos de comunidade tradicional
# 1 = Sim, 2 = Não
map_comunidade_tradicional = {
    1.0: "sim",
    2.0: "nao"
}

if "codigocontemcomunidadetradicional" in df.columns:
    df["codigocontemcomunidadetradicional"] = df["codigocontemcomunidadetradicional"].map(map_comunidade_tradicional)
    df["codigocontemcomunidadetradicional"] = df["codigocontemcomunidadetradicional"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigocontemcomunidadetradicional"].value_counts())
else:
    print("Coluna 'codigocontemcomunidadetradicional' não encontrada no dataframe")
```

```python
# Remover coluna codigodosesvacina
# Justificativa: 36% dos valores são nulos (53.570 de 148.741 registros)
# e os valores são muito variados e de difícil compreensão (16 combinações diferentes
# com padrões inconsistentes como "1,2", "1,2,2003", "1,2,3,4", etc.)
if "codigodosesvacina" in df.columns:
    df = df.drop("codigodosesvacina", axis=1)
    print("Coluna 'codigodosesvacina' removida com sucesso!")
else:
    print("Coluna 'codigodosesvacina' não encontrada no dataframe")
```

```python
# Mapear códigos de estado do teste 1
# 1 = Solicitado, 2 = Concluído, 3 = Coletado, 4 = Não Solicitado
map_estado_teste = {
    1.0: "solicitado",
    2.0: "concluido",
    3.0: "coletado",
    4.0: "nao_solicitado"
}

if "codigoestadoteste1" in df.columns:
    df["codigoestadoteste1"] = df["codigoestadoteste1"].map(map_estado_teste)
    df["codigoestadoteste1"] = df["codigoestadoteste1"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigoestadoteste1"].value_counts())
else:
    print("Coluna 'codigoestadoteste1' não encontrada no dataframe")
```

```python
# Mapear códigos de tipo de teste 1
# 1=RT-PCR, 2=RT LAMP, 3=Teste rápido antígeno, 4=Teste rápido IgM, 
# 5=Teste rápido IgG, 6=Teste sorológico IgA, 7=Teste sorológico IgM, 
# 8=Teste sorológico IgG, 9=Teste sorológico anticorpos totais
map_tipo_teste = {
    1.0: "rt_pcr",
    2.0: "rt_lamp",
    3.0: "teste_rapido_antigeno",
    4.0: "teste_rapido_igm",
    5.0: "teste_rapido_igg",
    6.0: "teste_sorologico_iga",
    7.0: "teste_sorologico_igm",
    8.0: "teste_sorologico_igg",
    9.0: "teste_sorologico_anticorpos_totais"
}

if "codigotipoteste1" in df.columns:
    df["codigotipoteste1"] = df["codigotipoteste1"].map(map_tipo_teste)
    df["codigotipoteste1"] = df["codigotipoteste1"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigotipoteste1"].value_counts())
else:
    print("Coluna 'codigotipoteste1' não encontrada no dataframe")
```

```python
# Mapear códigos de fabricante do teste 1
# Mapeamento apenas dos 3 principais fabricantes (representam 99,99% dos dados)
# Os demais códigos (322, 421, 841, 665, 632) possuem poucos registros (total de 11)
# e não interferem na integridade dos dados
map_fabricante_teste = {
    715.0: "ibmp",  # Instituto de Biologia Molecular do Paraná
    915.0: "outras_marcas",
    792.0: "bio_manguinhos_fiocruz"
}

if "codigofabricanteteste1" in df.columns:
    df["codigofabricanteteste1"] = df["codigofabricanteteste1"].map(map_fabricante_teste)
    df["codigofabricanteteste1"] = df["codigofabricanteteste1"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigofabricanteteste1"].value_counts())
else:
    print("Coluna 'codigofabricanteteste1' não encontrada no dataframe")
```

```python
# Mapear códigos de resultado do teste 1
# 1 = Reagente/Detectável (Positivo)
# 2 = Não Reagente/Não detectável (Negativo)
# 9 = Inválido/Inconclusivo ou Indeterminado
map_resultado_teste = {
    1.0: "reagente_detectavel",
    2.0: "nao_reagente_nao_detectavel",
    9.0: "invalido_inconclusivo"
}

if "codigoresultadoteste1" in df.columns:
    df["codigoresultadoteste1"] = df["codigoresultadoteste1"].map(map_resultado_teste)
    df["codigoresultadoteste1"] = df["codigoresultadoteste1"].fillna("ignorado")
    print("Mapeamento aplicado com sucesso!")
    print(df["codigoresultadoteste1"].value_counts())
else:
    print("Coluna 'codigoresultadoteste1' não encontrada no dataframe")
```

```python
##tratar coluna de datas
date_cols = [
    "datanotificacao", "datainiciosintomas", "dataencerramento",
    "dataprimeiradose", "datasegundadose",
    "datacoletateste1", "datacoletateste2",
    "datacoletateste3", "datacoletateste4"
]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
```

```python
df["idade"] = pd.to_numeric(df["idade"], errors="coerce")
```

```python
# Criar raw_line preservando acentuação
df["raw_line"] = df.apply(lambda x: ";".join(x.astype(str)), axis=1)
```

```python
# Salvar com encoding UTF-8 com BOM para melhor compatibilidade
import os

arquivo_csv = "dataset_limpo.csv"

# Remover arquivo anterior se existir
if os.path.exists(arquivo_csv):
    os.remove(arquivo_csv)
    print(f"Arquivo '{arquivo_csv}' anterior removido.")

# Salvar novo arquivo
df.to_csv(arquivo_csv, index=False, encoding="utf-8-sig", sep=";")
print(f"Novo arquivo '{arquivo_csv}' criado com sucesso!")
```
